## (ii)

Given that the gaussians are centered independently in the two coordinates and are seperable

$$\phi_A (x, y) = \exp \left[ - \alpha \left((x - x_A)^2 + (y - y_A)^2 \right) \right] = e^{- \alpha (x - x_A)^2 } e^{- \alpha (y - y_A)^2 }$$

The matrix elements of the coupling term can be seperated

$$G_{AB} = \langle \phi_A | x^4y^4 | \phi_B \rangle = \langle \phi_A (x) | x^4 | \phi_B (x) \rangle \langle \phi_A (y) | y^4 | \phi_B (y)\rangle$$

The analytical integral from of the generating function can be substituted, giving the matrix elements

$$\langle \phi_A (x) | x^4 | \phi_B (x) \rangle \langle \phi_A (y) | y^4 | \phi_B (y)\rangle = g(x_A, x_B) \cdot g(y_A, y_B)$$

$$ G_{AB} = \left( s(x_A, x_B) \left[ \frac{3}{16\alpha^2} + \frac{3}{8\alpha}(x_A + x_B)^2 + \frac{1}{16}(x_A + x_B)^4 \right] \right) \left( s(y_A, y_B) \left[ \frac{3}{16\alpha^2} + \frac{3}{8\alpha}(y_A + y_B)^2 + \frac{1}{16}(y_A + y_B)^4 \right] \right)$$

$$ G_{AB} = S_{AB} \left( \left[ \frac{3}{16\alpha^2} + \frac{3}{8\alpha}(x_A + x_B)^2 + \frac{1}{16}(x_A + x_B)^4 \right] \left[ \frac{3}{16\alpha^2} + \frac{3}{8\alpha}(y_A + y_B)^2 + \frac{1}{16}(y_A + y_B)^4 \right]\right)$$

where

$$s(x_A, x_B) = \sqrt{\frac{\pi}{2 \alpha}} \exp \left[ -\frac{\alpha}{2} (x_A - x_B)^2 \right] \quad s(y_A, y_B) = \sqrt{\frac{\pi}{2 \alpha}} \exp \left[ -\frac{\alpha}{2} (y_A - y_B)^2 \right] \quad S_{AB} = s(x_A, x_B) \cdot s(y_A, y_B)$$

## (iii) - (iv)

```python
def construct_K(n, spacing=1):
    K = np.zeros(((2*n+1)**2, 2))
    count = 0
    for i in range(-n,n+1):
        for j in range(-n,n+1):
            K[count,:] = [i*spacing, j*spacing]
            count += 1
    return K
```

```python
def S_AB_func(xA, yA, xB, yB, alpha):
    return (0.5*np.pi/alpha) * np.exp(-0.5*alpha* (xA - xB)**2 - 0.5*alpha* (yA - yB)**2)

def H_AB_func(xA, yA, xB, yB, alpha):
    return 0.5* S_AB_func(xA, yA, xB, yB, alpha) * (2*alpha + 1/(2*alpha) - alpha**2 * (xA - xB)**2 - alpha**2 * (yA - yB)**2 \
                                                    + 0.25*((xA + xB)**2 ) + 0.25*((yA + yB)**2 ))

def G_AB_func(xA, yA, xB, yB, alpha):
    return S_AB_func(xA, yA, xB, yB, alpha) * (3/(16*alpha**2) + 3/(8*alpha)*(xA + xB)**2 + 1/16*(xA + xB)**4) \
                                                    * (3/(16*alpha**2) + 3/(8*alpha)*(yA + yB)**2 + 1/16*(yA + yB)**4)

def S_H_G_AB_func(xy_A, xy_B, alpha):
    xA, yA = xy_A.T
    xB, yB = xy_B.T
    S_AB = S_AB_func(xA, yA, xB, yB, alpha)
    H_AB = H_AB_func(xA, yA, xB, yB, alpha)
    G_AB = G_AB_func(xA, yA, xB, yB, alpha)
    return S_AB, H_AB, G_AB

def H_AB_int_func(H_AB, G_AB, a):
    return H_AB + (a * G_AB)
```

```python
def find_eigval_eigvec(S_AB, H_AB):
    S_inv_H = np.linalg.inv(S_AB) @ H_AB
    eigval, eigvec = np.linalg.eig(S_inv_H)
    
    index_sort = np.argsort(eigval)
    eigval_sorted = eigval[index_sort]
    eigvec_sorted = eigvec[:, index_sort]
    
    for i in range(len(eigvec_sorted)):
        eigvec_sorted[:,i] /= np.sqrt(eigvec_sorted[:,i] @ S_AB @ eigvec_sorted[:,i])
        
    return eigval_sorted.real, eigvec_sorted
```

## (v)

```python
alpha = 2

K = construct_K(7, 0.5)
S_AB, H_AB, G_AB = S_H_G_AB_func(K[:,None], K[None,:], alpha)

eigval, eigvec = find_eigval_eigvec(S_AB, H_AB)
print(f'Ground State (a = 0): {eigval[0]}')

a = 0.005
H_AB_int = H_AB_int_func(H_AB, G_AB, a)
eigval, eigvec = find_eigval_eigvec(S_AB, H_AB_int)
print(f'Ground State (a = 0.005): {eigval[0]}')
```

```python
a_vals = [0.001, 0.01, 0.05, 0.1, 0.5, 1, 10]
E_0s = []
E_PTs = []
errors = []

for a in a_vals:
    H_AB_int = H_AB_int_func(H_AB, G_AB, a)
    eigval, eigvec = find_eigval_eigvec(S_AB, H_AB_int)
    # print(f'Ground State (a = {a}): {eigval[0]}')
    E_0s.append("{:.8f}".format(eigval[0]))
    E_PT = 1 + (9*a/16)
    E_PTs.append(E_PT)
    error = (E_PT - eigval[0]) / eigval[0]
    errors.append("{:.8f}".format(error))

print(E_0s)
print(E_PTs)
print(errors)
```

```python
def gaussian(x, y, x_A, y_A, alpha=2.):
    x_A = np.asarray(x_A)[:, None, None]
    y_A = np.asarray(y_A)[:, None, None]
    return np.exp(-alpha*(x - x_A) ** 2. - alpha*(y - y_A) ** 2.)

def psi(K, eigvec, xgrid, ygrid):
    x_A, y_A = K.T
    gaussians = gaussian(xgrid, ygrid, x_A, y_A)
    return np.tensordot(eigvec, gaussians, axes=(0, 0))
```

```python
warnings.filterwarnings("ignore")

x_space, y_space = np.arange(-2.5,2.5,0.1), np.arange(-2.5,2.5,0.1) 
xgrid, ygrid = np.meshgrid(x_space, y_space)

H_AB_int = H_AB_int_func(H_AB, G_AB, 1)
eigval, eigvec = find_eigval_eigvec(S_AB, H_AB_int)
eval = psi(K, eigvec[:,0], xgrid, ygrid)


plt.rcParams['text.usetex'] = True

fig, axes = plt.subplots(1,2,figsize=(12, 6))
axes[0].contourf(xgrid, ygrid, eval, levels=20, cmap=cm.seismic)
axes[0].axis('equal') 
axes[0].set_title(r'$\psi_0 (x,y)$ contour $a=1$', fontsize=20)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

fig.delaxes(axes[1])
ax3d = fig.add_subplot(1, 2, 2, projection='3d')
ax3d.plot_surface(xgrid, ygrid, eval, cmap=cm.seismic)
ax3d.set_title(r'$\psi_0 (x,y)$ surface $a=1$', fontsize=20)
ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')

plt.savefig(f'Visualizations/1_viii.png')
```

The interaction term $ax^4y^4$ in the interacting hamiltonian has a steep barrier everywhere except the x and y axes, allowing more of the wavefunction to exist along these axes.

```python
from numpy import *
from numpy.linalg import *
from numpy.random import *
from matplotlib.pyplot import *
from scipy.special import binom, erf, erfc

def eigSinvH(S,H): # compute energies and coefficients
    SinvH = inv(S) @ H
    E, U = eig(SinvH)
    order = argsort(real(E))
    c = zeros((K, K))
    for i in range(K):
        c[:, i] = real(U[:, order[i]])
        c[:, i] = c[:, i] / sqrt(c[:, i] @ S @ c[:, i])
    E = sort(real(E))
    return E, c

def F0(x): # compute F_0 matrix for elec_nuc energy term
    if (x < 1e-8):
        return 1 - x/3
    else:
        return 0.5*sqrt(pi)*erf(sqrt(x))/sqrt(x)
    
def compute_overlap(alpha, beta, RA, RB): # Compute analytical overlap matrix elements
    absum = alpha + beta 
    abfac = alpha * beta / absum
    dRAB = RA - RB # shape: (5,5,3) = (5,5,3) - (5,5,3) <-- reshape (5,1,3) - (1,5,3)
    dRAB2 = dRAB @ dRAB # shape: (5,5) = (5,5,3) @ (5,5,3)
    return (pi/absum)**(3/2) * exp(-abfac * dRAB2)

def compute_kinetic_energy(alpha, beta, RA, RB): # Compute analytical kinetic energy matrix elements
    absum = alpha + beta
    abfac = alpha * beta / absum
    dRAB = RA - RB
    dRAB2 = dRAB @ dRAB
    return (pi/absum)**(3/2) * exp(-abfac * dRAB2) * \
            abfac * (3 - 2*abfac * dRAB2)

def compute_elec_nuc_energy(alpha, beta, RA, RB, RC): # Compute analytical elec_nuc energy matrix elements
    absum = alpha + beta
    abfac = alpha * beta / absum
    dRAB = RA - RB
    dRAB2 = dRAB @ dRAB
    RP = (alpha*RA + beta*RB)/absum
    dRPC = RP - RC
    dRPC2 = dRPC @ dRPC
    return -(2*pi/absum)*exp(-abfac * dRAB2) * \
            F0(absum * dRPC2)

K = 3 # number of basis functions 
L = 2 # number of Gaussians used to build each basis function

widths = zeros((L,K)) 
contraction_coeffs = zeros((L,K)) # we called these d
centers = zeros((L,K,3))

widths[:,0] = [0.6, 0.6] # set gaussian widths for each gaussian pair
widths[:,1] = [0.1, 0.1]
widths[:,2] = [0.02, 0.02]

contraction_coeffs[:,0] = [1, -1] # set contraction coefficients
contraction_coeffs[:,1] = [1, -1]
contraction_coeffs[:,2] = [1, -1]

offset = 0.01 # we called this Delta
centers[:,0,2] = [offset, -offset] # set offset in the z axis for all gaussians
centers[:,1,2] = [offset, -offset]
centers[:,2,2] = [offset, -offset]

R_nucleus = zeros(3) # nucleus at origin

S = zeros((K,K))
T = zeros((K,K))
U = zeros((K,K))

for mu in range(K): # iterate over each basis functions index mu
    for A in range(L): # iterate over each gaussians under basis mu
        alpha = widths[A,mu] # collect gaussian width, contraction coefficient, and coordiates
        dAmu = contraction_coeffs[A,mu]
        RA = centers[A,mu,:]

        for nu in range(K): # iterate over each basis functions index nu
            for B in range(L): # iterate over each gaussians under basis nu
                beta = widths[B,nu]
                dBnu = contraction_coeffs[B,nu]
                RB = centers[B,nu,:]

                # construction of overlap, kinetic, and potential matricies
                S[mu,nu] += dAmu * dBnu * compute_overlap(alpha, beta, RA, RB)
    
                T[mu,nu] += dAmu * dBnu * compute_kinetic_energy(alpha, beta, RA, RB)
    
                U[mu,nu] += dAmu * dBnu * compute_elec_nuc_energy(alpha, beta, RA, RB, R_nucleus)
    
H = T + U
E, c = eigSinvH(S,H) # Energy and coefficients of diagonalized wavefunction

print(E) 

x_values = arange(-10,10,0.1) # defining grid
z_values = x_values
number_of_values = size(x_values)
state = 0 # picking the p_z orbital
psi = zeros((number_of_values,number_of_values)) # wavefunction evaluation grid
    
for i in range(number_of_values): # iterate over all grid points
    x = x_values[i]
    for j in range(number_of_values):
        z = z_values[j]

        for mu in range(K): # iterate over all basis functions
            for A in range(L): # iterate over each gaussian for each basis function
                alpha = widths[A, mu]
                dAmu = contraction_coeffs[A, mu]
                RA = centers[A, mu, :]
                xA = RA[0]
                zA = RA[2]
                phiA = exp(-alpha * ( (x-xA)**2 + (z-zA)**2  ) ) # construct gaussian basis with given parameters
    
                psi[i,j] += c[mu,state] * dAmu * phiA # sum over all gaussian basis with weights of contraction_coeff and eigenstate coeff

clf()
contourf(x_values, z_values, psi, levels=20, cmap=cm.seismic)
axis('equal')
```

## (ii)

Four gaussian functions would be needed for each lobe due to the inversion symmetry of the opposite lobes. These gaussians need to be off-centered in two seperate axes for $d_{xy}$, $d_{xz}$, and $d_{yz}$ orbitals because of dihedral lobes. For the $d_{xz}$ the following arrangement of contraction coefficients and centers offsets are needed: 

| d-orbital type | $d = 1$ (pos contract_coeff) | $d = -1$ (neg contract_coeff) |
| --- | --- | --- |
| $d_{xz}$ | [$-\delta, 0, +\delta$] [$+\delta, 0, -\delta$] | [$+\delta, 0, +\delta$] [$-\delta, 0, -\delta$] |

Where $\delta$ are the offsets and $d$ is the contraction coefficient.

```python
S = np.zeros((K,K))
T = np.zeros((K,K))
U = np.zeros((K,K))

for mu in range(K):
    for A in range(L):
        alpha = widths[A,mu]
        dAmu = contraction_coeffs[A,mu]
        RA = centers[A,mu,:]

        for nu in range(K):
            for B in range(L):
                beta = widths[B,nu]
                dBnu = contraction_coeffs[B,nu]
                RB = centers[B,nu,:]

                S[mu,nu] += dAmu * dBnu * compute_overlap(alpha, beta, RA, RB)
                T[mu,nu] += dAmu * dBnu * compute_kinetic_energy(alpha, beta, RA, RB)
                U[mu,nu] += dAmu * dBnu * compute_elec_nuc_energy(alpha, beta, RA, RB, R_nucleus)

H = T + U
E, c = eigSinvH(S,H)

print(E[0])
```
