# Question 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
```

## (i)

```python
A = np.array([[3, -0.5, 0, 0, 0],
              [-0.5, 1.5, -0.5, 0, 0],
              [0, -0.5, 1, -0.5, 0],
              [0, 0, -0.5, 1.5, -0.5,],
              [0, 0, 0, -0.5, 3]])
```

## (ii)

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
eigenvectors = eigenvectors.transpose()
eigenvalues, eigenvectors
```

## (iii) and (iv)

```python
E_max = eigenvalues.max()
c_max = eigenvectors[np.where(np.isclose(eigenvalues, E_max))][0]
E_min = eigenvalues.min()
c_min = eigenvectors[np.where(np.isclose(eigenvalues, E_min))][0]

print('c_max =', c_max, 'has eigenvalue', E_max)
print('c_min =', c_min, 'has eigenvalue', E_min)
```

## (v)

```python
print(f"A @ c_max: {A @ c_max}")
print(f"E_max * c_max: {E_max * c_max}")
print(f"\nA @ c_min: {A @ c_min}")
print(f"E_min * c_min: {E_min * c_min}")
```

## (vi) and (vii)

```python
print(f"c_max @ c_max is: {c_max @ c_max}")
print(f"c_min @ c_min is: {c_min @ c_min}")

print(f"c_min @ c_max is: {c_min @ c_max}")
```

# Question 3

```python
def x(t):
    return np.sin(t) * (np.exp(2 * np.cos(t/2)) - 2 * np.cos(8*t) - (np.sin(t/6))**6)

def y(t):
    return np.cos(t) * (np.exp(2 * np.cos(t/2)) - 2 * np.cos(8*t) - (np.sin(t/6))**6)

def circle(r, t, x, y):
    return [r*np.cos(t) + x, r*np.sin(t) + y]

def curve(x):
    return 1 + 0.1 * (x)**2

def censor_bar(plt, y):
    return plt.axhline(y=y, xmin=0.3, xmax=0.7, color='black', linewidth=40)
```

```python
t = np.linspace(0, 12*np.pi, 10000)
curve_x = np.linspace(-3, 3, 1000)

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

plt.figure(figsize=(8,8))
plt.plot(x(t), y(t), label='x')
plt.plot(circle(0.3, t, -2, 6)[0], circle(0.3, t, -2, 6)[1], color=colors[0])
plt.plot(circle(0.3, t, 2, 6)[0], circle(0.3, t, 2, 6)[1], color=colors[0])
plt.plot(curve_x, curve(curve_x), linewidth=10, color=colors[3])
plt.xlabel('x')
plt.ylabel('y')
plt.title('very large amoeba holding flower')
```

```python
t = np.linspace(0, 12*np.pi, 10000)
curve_x = np.linspace(-3, 3, 1000)

plt.figure(figsize=(8,8))
plt.plot(x(t), y(t), label='x')
plt.plot(circle(0.3, t, -2, 6)[0], circle(0.3, t, -2, 6)[1], color=colors[0])
plt.plot(circle(0.3, t, 2, 6)[0], circle(0.3, t, 2, 6)[1], color=colors[0])
plt.plot(curve_x, curve(curve_x), linewidth=10, color=colors[3])
censor_bar(plt, 6)
plt.xlabel('x')
plt.ylabel('y')
plt.title('very large censored amoeba holding flower')
```

# Question 4

```python
def f(x):
    return 0.5 * x**2

def g_1(x):
    return 0.5 + 0.5 * np.exp(-1 * x**2 / 2)
    
def g_2(x):
    return 1.5 + 0.5 * np.exp(-1 * x**2 / 2)
    
def g_3(x):
    return 2.5 + 0.5 * (2 * x**2 - 1) * np.exp(-1 * x**2 / 2)
```

```python
x = np.linspace(-3, 3, 1000)

plt.figure()
plt.plot(x, f(x))
plt.axhline(y=0.5, color='black')
plt.axhline(y=1.5, color='black')
plt.axhline(y=2.5, color='black')
plt.plot(x, g_1(x))
plt.plot(x, g_2(x))
plt.plot(x, g_3(x))
```
