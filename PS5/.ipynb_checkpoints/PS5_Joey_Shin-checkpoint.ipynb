```python
# import packages for basic math, plotting, linear algebra, etc.
from numpy import *
from numpy.linalg import *
from numpy.random import *
from matplotlib.pyplot import *
from scipy.special import binom, erf, erfc

# Hamiltonian diagonalization method initialization 
def eigSinvH(S,H):
    SinvH = inv(S) @ H
    E, U = eig(SinvH)

    order = argsort(E)
    c = zeros((K, K))
    for i in range(K):
        c[:, i] = U[:, order[i]]
        c[:, i] = c[:, i] / sqrt(c[:, i] @ S @ c[:, i])

    E = sort(E)

    return E, c

# gaussian basis exponents & grid
alpha = 2
deltax = 0.5

n = 10
K = 2*n + 1

center = arange(-n*deltax,(n+1)*deltax,deltax)

S = zeros((K,K))
h = zeros((K,K))
G = zeros((K,K))

# matrix elements construction
for A in range(K):
    xA = center[A]
    for B in range(K):
        xB = center[B]

        S[A,B] = sqrt(0.5*pi/alpha) * exp(-0.5*alpha* (xA - xB)**2 )

        h[A,B] = 0.5*S[A,B] * (alpha - alpha**2 * (xA - xB)**2 + \
                             0.25*(1/alpha + (xA + xB)**2 ))

        G[A,B] = S[A,B] * ( 3/(16*alpha**2) + \
                            (3/(8*alpha)) * (xA + xB)**2 + \
                            (1/16) * (xA + xB)**4  )

# Hamiltonian diagonalization, coefficient construction
E, c = eigSinvH(S,h)
c1 = c[:,0]
c2 = c[:,2] # using 3rd excited state as initial guess

a = 1

# SCF iterations
niterations = 100
for iteration in range(niterations):
    avx4 = c1 @ G @ c1 # evaluate <x^4> matrix
    avy4 = c2 @ G @ c2 # evaluate of <y^4> matrix

    heffx = h + a * avy4 * G # evaluate effective hamiltonian matrix
    heffy = h + a * avx4 * G

    # find new coefficients
    E, c = eigSinvH(S,heffx)
    c1 = c[:,0]
    e1 = E[0]

    E, c = eigSinvH(S,heffy)
    c2 = c[:,0]
    e2 = E[0]

    # compute new total energy
    Etot = e1 + e2 - a * avx4 * avy4
    print(Etot)

# Exact energy for a=1 is about 1.14

xvals = arange(-2.5,2.5,0.1) # define grid
yvals = xvals
nvals = size(xvals)

chi1 = 0 * xvals
chi2 = 0 * yvals

for A in range(K): # evaluate wavefunction on grid
    chi1 += c1[A] * exp(-alpha * (xvals - center[A])**2 )
    chi2 += c2[A] * exp(-alpha * (xvals - center[A])**2 )

# plot(xvals,chi1,'o')
# plot(yvals,chi2)

clf()
psi = zeros((nvals,nvals))

for i in range(nvals):
    x = xvals[i]
    for j in range(nvals):
        y = yvals[j]

        # evaluate total wavefunction
        psi[i,j] = chi1[i] * chi2[j]

clf()
contourf(xvals, yvals, psi, levels=20, cmap=cm.seismic)
axis('equal')
```

## (ii)

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

from tqdm import tqdm
import multiprocess as mp
from multiprocess import cpu_count
```

```python
def construct_K(n, spacing=1):
    K = np.zeros(((2*n+1)**2, 2))
    count = 0
    for i in range(-n,n+1):
        for j in range(-n,n+1):
            K[count,:] = [i*spacing, j*spacing]
            count += 1
    return K

def S_AB_func(xA, yA, xB, yB, alpha):
    return (0.5*np.pi/alpha) * np.exp(-0.5*alpha* (xA - xB)**2 - 0.5*alpha* (yA - yB)**2)

def H_AB_func(xA, yA, xB, yB, alpha):
    return 0.5* S_AB_func(xA, yA, xB, yB, alpha) * (2*alpha + 1/(2*alpha) - alpha**2 * (xA - xB)**2 - alpha**2 * (yA - yB)**2 \
                                                    + 0.25*((xA + xB)**2 ) + 0.25*((yA + yB)**2 ))

def G_AB_func(xA, yA, xB, yB, alpha):
    return S_AB_func(xA, yA, xB, yB, alpha) * (3/(16*alpha**2) + 3/(8*alpha)*(xA + xB)**2 + 1/16*(xA + xB)**4) \
                                                    * (3/(16*alpha**2) + 3/(8*alpha)*(yA + yB)**2 + 1/16*(yA + yB)**4)

def S_H_AB_func(xy_A, xy_B, alpha):
    xA, yA = xy_A.T
    xB, yB = xy_B.T
    S_AB =  S_AB_func(xA, yA, xB, yB, alpha)
    H_AB = H_AB_func(xA, yA, xB, yB, alpha)
    return S_AB, H_AB

def S_H_G_AB_func(xy_A, xy_B, alpha):
    xA, yA = xy_A.T
    xB, yB = xy_B.T
    S_AB = S_AB_func(xA, yA, xB, yB, alpha)
    H_AB = H_AB_func(xA, yA, xB, yB, alpha)
    G_AB = G_AB_func(xA, yA, xB, yB, alpha)
    return S_AB, H_AB, G_AB

def H_AB_int_func(H_AB, G_AB, a):
    return H_AB + (a * G_AB)

def find_eigval_eigvec(S_AB, H_AB):
    S_inv_H = np.linalg.inv(S_AB) @ H_AB
    eigval, eigvec = np.linalg.eig(S_inv_H)
    
    index_sort = np.argsort(eigval)
    eigval_sorted = eigval[index_sort]
    eigvec_sorted = eigvec[:, index_sort]
    
    for i in range(len(eigvec_sorted)):
        eigvec_sorted[:,i] /= np.sqrt(eigvec_sorted[:,i] @ S_AB @ eigvec_sorted[:,i])
        
    return eigval_sorted.real, eigvec_sorted
```

```python
aa = np.arange(0, 0.01, 0.0001)
E_MFs = []
E_PTs = []

for a in tqdm(aa):
    E, c = eigSinvH(S,h)
    c1 = c[:,0]
    c2 = c[:,2]

    niterations = 100
    for iteration in range(niterations):
        avx4 = c1 @ G @ c1 
        avy4 = c2 @ G @ c2 
    
        heffx = h + a * avy4 * G 
        heffy = h + a * avx4 * G
    
        E, c = eigSinvH(S,heffx)
        c1 = c[:,0]
        e1 = E[0]
    
        E, c = eigSinvH(S,heffy)
        c2 = c[:,0]
        e2 = E[0]
    
    E_MFs.append(e1 + e2 - a * avx4 * avy4)
    E_PT = 1 + (9*a/16)
    E_PTs.append(E_PT)
```

```python
plt.rcParams['text.usetex'] = True

fig, ax = plt.subplots()
ax.plot(aa, E_MFs, label=r'$E_{0}^{MF}$')
ax.plot(aa, E_PTs, label=r'$E_{0}^{PT}$')
ax.set_title(r'$E_{0}^{MF}$ vs $E_{0}^{PT}$', fontsize=15)
ax.set_xlabel(r'a', fontsize=15)
ax.set_ylabel(r'$E$', fontsize=15)
ax.legend()
plt.savefig('Visualizations/1_ii.png')
```

The mean field energy gives a variationally lower energy as a increases compared to pertubation theory, which already had a small error incomparison to the true energy at small interactions. 

## (iii)

```python
a_vals = [0.001, 0.01, 0.05, 0.1, 0.5, 1, 10]
E_0s = []
E_MFs = []
E_PTs = []
errors_PT = []
errors_MF = []

for a in tqdm(a_vals):
    # Basis Expansion
    K_grid = construct_K(7, 0.5)
    S_AB, H_AB, G_AB = S_H_G_AB_func(K_grid[:,None], K_grid[None,:], alpha)    
    H_AB_int = H_AB_int_func(H_AB, G_AB, a)
    eigval, eigvec = find_eigval_eigvec(S_AB, H_AB_int)
    E_0s.append("{:.8f}".format(eigval[0]))
    
    # Mean Field Theory
    E, c = eigSinvH(S,h)
    c1 = c[:,0]
    c2 = c[:,0]
    
    niterations = 100
    for iteration in range(niterations):
        avx4 = c1 @ G @ c1 
        avy4 = c2 @ G @ c2 
    
        heffx = h + a * avy4 * G 
        heffy = h + a * avx4 * G
    
        E, c = eigSinvH(S,heffx)
        c1 = c[:,0]
        e1 = E[0]
    
        E, c = eigSinvH(S,heffy)
        c2 = c[:,0]
        e2 = E[0]
    E_MF = e1 + e2 - a * avx4 * avy4
    E_MFs.append("{:.8f}".format(E_MF))

    # Perturbation Theory
    E_PT = 1 + (9*a/16)
    E_PTs.append(E_PT)

    # Perturbation Theory Error
    error_PT = (E_PT - eigval[0]) / eigval[0]
    errors_PT.append("{:.8f}".format(error_PT))

    # Mean Field Theory Error
    error_MF = (E_MF - eigval[0]) / eigval[0]
    errors_MF.append("{:.8f}".format(error_MF))

print(f'Basis Expansion: {E_0s}')
print(f'Mean Field Theory: {E_MFs}')
print(f'Perturbation Theory:  {E_PTs}')
print(f'Mean Field Theory Errors: {errors_MF}')
print(f'Perturbation Theory Errors: {errors_PT}')
```

## (i)

$$| x_A, y_A \rangle = \phi_A (x, y)$$

$$ \hat{P} \phi_A (x, y) = \phi_A (y, x)$$

$$\hat{P} \exp[-\alpha(x-x_A)^2 -\alpha(y-y_A)^2] = \exp[-\alpha(y-x_A)^2 -\alpha(x-y_A)^2] \neq - \exp[-\alpha(x-x_A)^2 -\alpha(y-y_A)^2]$$

$$ \hat{P} \phi_A (x, y) \neq -\phi_A (x, y)$$

## (ii)

Applying the antisymmetrizer on the wavefunction and expanding out the gaussian:

$$\hat{A} | x_A, y_A \rangle = \hat{A} \phi_A (x, y) = \frac{1}{N!} \sum_{j}^{N!} (-1)^{P_j} \hat{P}  \phi_A (x, y) = \frac{1}{2!} \left[  \phi_A (x, y) -  \phi_A (y, x) \right]$$

$$\frac{1}{2!} \left[  \phi_A (x, y) -  \phi_A (y, x) \right] = \frac{1}{2!} \left( \exp[-\alpha(x-u_A)^2 -\alpha(y-v_A)^2] - \exp[-\alpha(y-u_A)^2 -\alpha(x-v_A)^2] \right)$$

Rearranging the terms and applying the permutation operator

$$\frac{1}{2!} \left( \exp[-\alpha(x-u_A)^2 -\alpha(y-v_A)^2] - \exp[-\alpha(x-v_A)^2 -\alpha(y-u_A)^2] \right) $$

Removing the normalization constant and applying the projection operator: 

$$\hat{P} \exp[-\alpha(x-u_A)^2 -\alpha(y-v_A)^2] - \exp[-\alpha(x-v_A)^2 -\alpha(y-u_A)^2] = \exp[-\alpha(y-u_A)^2 -\alpha(x-v_A)^2] - \exp[-\alpha(y-v_A)^2 -\alpha(x-u_A)^2]$$

$$ = \exp[-\alpha(y-u_A)^2 -\alpha(x-v_A)^2] - \exp[-\alpha(y-v_A)^2 -\alpha(x-u_A)^2] = -\left( \exp[-\alpha(x-u_A)^2 -\alpha(y-v_A)^2] - \exp[-\alpha(x-v_A)^2 -\alpha(y-u_A)^2] \right)$$

$$ \hat{P} \phi_{A} (x, y) = -\phi_{A} (x, y)$$

```python
def gaussian(x, y, x_A, y_A, alpha=2.):
    return np.exp(-alpha*(x - x_A) ** 2. - alpha*(y - y_A) ** 2.) - np.exp(-alpha*(x - y_A) ** 2. - alpha*(y - x_A) ** 2.)

x_space, y_space = np.arange(-4,4,0.1), np.arange(-4,4,0.1) 
xgrid, ygrid = np.meshgrid(x_space, y_space)
eval = gaussian(xgrid, ygrid, 2, 1)

plt.rcParams['text.usetex'] = True

fig, ax = plt.subplots()
ax.contourf(xgrid, ygrid, eval, levels=20, cmap=cm.seismic)
ax.axis('equal') 
ax.set_title(r'$\phi_A (x,y)$', fontsize=20)
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.savefig(f'Visualizations/2_ii.png')
```

## (iii)

$$| \phi_A \rangle = | u_A, v_A \rangle - | v_A, u_A \rangle$$

$$S_{AB} = \langle \phi_A | \phi_B \rangle = \left[ \langle u_A, v_A | - \langle v_A, u_A | \right] \left[ | u_B, v_B \rangle - | v_B, u_B \rangle \right]$$

$$S_{AB} = \langle u_A, v_A | u_B, v_B \rangle - \langle u_A, v_A | v_B, u_B \rangle - \langle v_A, u_A | u_B, v_B \rangle + \langle v_A, u_A | v_B, u_B \rangle$$

Seperating each basis in the overlap: 

$$\langle u_A, v_A | u_B, v_B \rangle = \int dydx \, \left[ e^{-\alpha (x - u_A)^2} e^{-\alpha (y - v_A)^2} \right] \left[ e^{-\alpha (x - u_B)^2} e^{-\alpha (y - v_B)^2} \right] = \int dx \, e^{-\alpha (x - u_A)^2} e^{-\alpha (x - u_B)^2} \int dy \, e^{-\alpha (y - v_A)^2} e^{-\alpha (y - v_B)^2} = \langle u_A | u_B \rangle \langle v_A | v_B \rangle$$


Solving for each expression:

$$\langle u_A, v_A | u_B, v_B \rangle = \langle u_A | u_B \rangle \langle v_A | v_B \rangle = s(u_A, u_B) s(v_A, v_B)$$
$$\langle u_A, v_A | v_B, u_B \rangle = s(u_A, v_B) s(v_A, u_B)$$
$$\langle v_A, u_A | u_B, v_B \rangle = s(v_A, u_B) s(u_A, v_B)$$
$$\langle v_A, u_A | v_B, u_B \rangle = s(v_A, v_B) s(u_A, u_B)$$


Collecting the exprssion and combining like terms:

$$S_{AB} = s(u_A, u_B) s(v_A, v_B) - s(u_A, v_B) s(v_A, u_B) - s(v_A, u_B) s(u_A, v_B) + s(v_A, v_B) s(u_A, u_B)$$

$$S_{AB} = 2 s(u_A, u_B) s(v_A, v_B) - 2 s(v_A, u_B) s(u_A, v_B) $$

## (iv)

$$ h_{AB} = \langle \phi_A | h(x) | \phi_B \rangle = \left( \langle u_A, v_A | - \langle v_A, u_A | \right) [h(x)] \left( | u_B, v_B \rangle - | v_B, u_B \rangle \right)$$

$$h_{AB} = \langle u_A, v_A | h(x) | u_B, v_B \rangle - \langle u_A, v_A | h(x) | v_B, u_B \rangle - \langle v_A, u_A | h(x) | u_B, v_B \rangle + \langle v_A, u_A | h(x) | v_B, u_B \rangle$$

Since $h(x)$ is dependent on $x$ basis: 

$$\langle u_A, v_A | h(x) | u_B, v_B \rangle = \int_{-\infty}^{\infty} dydx \, \left[ e^{-\alpha (x - u_A)^2} e^{-\alpha (y - v_A)^2} \right] h(x) \left[ e^{-\alpha (x - u_B)^2} e^{-\alpha (y - v_B)^2} \right] $$
$$ = \int_{-\infty}^{\infty} dx \, e^{-\alpha (x - u_A)^2} h(x) e^{-\alpha (x - u_B)^2} \int_{-\infty}^{\infty} dy \, e^{-\alpha (y - v_A)^2} e^{-\alpha (y - v_B)^2} = \langle u_A | h(x)| u_B \rangle \langle v_A | v_B \rangle$$

Defining:

$$f(x_A, x_B) = \int_{-\infty}^{\infty} dx \, e^{-\alpha (x - x_A)^2} h(x) e^{-\alpha (x - x_B)^2}$$

Solving for each expression and collecting the terms: 

$$\langle u_A, v_A | h(x) | u_B, v_B \rangle = \langle u_A | h(x) | u_B \rangle \langle v_A | v_B \rangle = f(u_A, u_B) s(v_A, v_B)$$
$$\langle u_A, v_A | h(x) | v_B, u_B \rangle = f(u_A, v_B) s(v_A, u_B)$$
$$\langle v_A, u_A | h(x) | u_B, v_B \rangle = f(v_A, u_B) s(u_A, v_B)$$
$$\langle v_A, u_A | h(x) | v_B, u_B \rangle = f(v_A, v_B) s(u_A, u_B)$$

$$h_{AB} = f(u_A, u_B) s(v_A, v_B) + f(v_A, v_B) s(u_A, u_B) - f(u_A, v_B) s(v_A, u_B) - f(v_A, u_B) s(u_A, v_B)$$

## (vi)

```python
def centers(n, deltax):
    K = n*(2*n + 1)
    center = zeros((K,2))
    count=0
    for i in range(-(n-1),n+1):
        uA = i*deltax
        for j in range(-n,i):
            vA = j*deltax
            center[count,:] = [uA, vA]
            count=count+1
    return center
```

```python
def s(uA,uB,alpha=2):
    return sqrt(0.5 * pi / alpha) * exp(-0.5 * alpha * (uA - uB) ** 2)

def f(uA,uB,alpha=2):
    return 0.5*sqrt(0.5 * pi / alpha) * exp(-0.5 * alpha * (uA - uB) ** 2) \
    * (alpha - alpha**2 * (uA - uB)**2 + \
    0.25*(1/alpha + (uA + uB)**2 ))

def g(uA,uB,alpha=2):
    return sqrt(0.5 * pi / alpha) * exp(-0.5 * alpha * (uA - uB) ** 2) \
    * (3 / (16 * alpha ** 2) + \
    (3 / (8 * alpha)) * (uA + uB) ** 2 \
    + (1 / 16) * (uA + uB) ** 4)
```

```python
n = 10
deltax = 0.5
alpha = 2

def S_AB_func(uA, vA, uB, vB, alpha):
    return 2*s(uA,uB)*s(vA,vB) - 2*s(vA,uB)*s(uA,vB)

def H_AB_func(uA, vA, uB, vB, alpha):
    return 2*(f(uA,uB)*s(vA,vB) + f(vA,vB)*s(uA,uB) - f(uA,vB)*s(vA,uB) - f(vA,uB)*s(uA,vB))

def G_AB_func(uA, vA, uB, vB, alpha):
    return 2*g(uA,uB)*g(vA,vB) - 2*g(vA,uB)*g(uA,vB)

center = centers(n, deltax)
S_AB, H_AB = S_H_AB_func(center[:,None], center[None,:], alpha)
eigval, eigvec = find_eigval_eigvec(S_AB, H_AB)
eigval[:9]
```

## (i)

$$ G_{AB} = \langle \phi_A | x^4 y^4 | \phi_B \rangle = \left( \langle u_A, v_A | - \langle v_A, u_A | \right) [x^4 y^4] \left( | u_B, v_B \rangle - | v_B, u_B \rangle \right)$$

$$G_{AB} = \langle u_A, v_A | x^4 y^4 | u_B, v_B \rangle - \langle u_A, v_A | x^4 y^4 | v_B, u_B \rangle - \langle v_A, u_A | x^4 y^4 | u_B, v_B \rangle + \langle v_A, u_A | x^4 y^4 | v_B, u_B \rangle$$

$$\langle u_A, v_A | x^4 y^4 | u_B, v_B \rangle = \int_{-\infty}^{\infty} dydx \, \left[ e^{-\alpha (x - u_A)^2} e^{-\alpha (y - v_A)^2} \right] x^4 y^4 \left[ e^{-\alpha (x - u_B)^2} e^{-\alpha (y - v_B)^2} \right] $$
$$ = \int_{-\infty}^{\infty} dx \, e^{-\alpha (x - u_A)^2} x^4 e^{-\alpha (x - u_B)^2} \int_{-\infty}^{\infty} dy \, e^{-\alpha (y - v_A)^2} y^4 e^{-\alpha (y - v_B)^2} = \langle u_A | x^4 | u_B \rangle \langle v_A | y^4 | v_B \rangle$$

Defining:

$$g(x_A, x_B) = \int_{-\infty}^{\infty} dx \, e^{-\alpha (x - x_A)^2} x^4 e^{-\alpha (x - x_B)^2}$$

Solving for each expression and collecting the terms: 

$$\langle u_A, v_A | x^4 y^4 | u_B, v_B \rangle = \langle u_A | x^4 | u_B \rangle \langle v_A | y^4 | v_B \rangle = g(u_A, u_B) g(v_A, v_B)$$
$$\langle u_A, v_A | x^4 y^4 | v_B, u_B \rangle = g(u_A, v_B) g(v_A, u_B)$$
$$\langle v_A, u_A | x^4 y^4 | u_B, v_B \rangle = g(v_A, u_B) g(u_A, v_B)$$
$$\langle v_A, u_A | x^4 y^4 | v_B, u_B \rangle = g(v_A, v_B) g(u_A, u_B)$$

$$G_{AB} = 2 g(u_A, u_B) g(v_A, v_B) - 2 g(v_A, u_B) g(u_A, v_B)$$

## (iii)

```python
def gaussian(x, y, x_A, y_A, alpha=2.):
    x_A = np.asarray(x_A)[:, None, None]
    y_A = np.asarray(y_A)[:, None, None]
    return np.exp(-alpha*(x - x_A) ** 2. - alpha*(y - y_A) ** 2.)

def psi(K, eigvec, xgrid, ygrid):
    x_A, y_A = K.T
    gaussians = gaussian(xgrid, ygrid, x_A, y_A) - gaussian(xgrid, ygrid, y_A, x_A)
    return np.tensordot(eigvec, gaussians, axes=(0, 0))

warnings.filterwarnings("ignore")

a_vals = [0, 0.001, 0.01, 0.1, 1, 10]
x_space, y_space = np.arange(-2.5,2.5,0.1), np.arange(-2.5,2.5,0.1) 
xgrid, ygrid = np.meshgrid(x_space, y_space)

for a in a_vals:
    H_AB_int = H_AB_int_func(H_AB, G_AB, a)
    eigval, eigvec = find_eigval_eigvec(S_AB, H_AB_int)
    eval = psi(center, eigvec[:,0], xgrid, ygrid)
    if eval[0][0] < 0.0:
        eval *= -1

    fig, axes = plt.subplots(1,2,figsize=(12, 6))
    axes[0].contourf(xgrid, ygrid, eval, levels=20, cmap=cm.seismic)
    axes[0].axis('equal') 
    axes[0].set_title(fr'$\psi_0 (x,y)$ contour $a={a}$', fontsize=20)
    axes[0].set_xlabel('X')
    axes[0].set_ylabel('Y')
    
    fig.delaxes(axes[1])
    ax3d = fig.add_subplot(1, 2, 2, projection='3d')
    ax3d.plot_surface(xgrid, ygrid, eval, cmap=cm.seismic)
    ax3d.set_title(fr'$\psi_0 (x,y)$ surface $a={a}$', fontsize=20)
    ax3d.set_xlabel('X')
    ax3d.set_ylabel('Y')
    
    plt.savefig(f'Visualizations/3_iii_{a}.png')
```
