$F_0(x) = \dfrac{\sqrt{\pi}}{2} \dfrac{erf(\sqrt{x})}{\sqrt{x}}$.

$F_0(x) \approx 1 - \dfrac{1}{3}x$ when $x \approx 0$

```python
import numpy as np
import scipy as sp
```

```python
from matplotlib import pyplot as plt
```

```python
def F0(x_list):
    result = []
    for x in x_list:
        if x < 1e-6:
            f0 = 1-x/3.
        else:
            f0 = 0.5*np.sqrt(np.pi)*sp.special.erf(np.sqrt(x))/np.sqrt(x)
        result.append(f0)
    return np.array(result)
```

```python
x = np.linspace(0,1,1000)
```

```python
mask = ( x < 0.5)
```

```python
def F0(x_list):
    result = 0.5*np.sqrt(np.pi)*sp.special.erf(np.sqrt(x))/np.sqrt(x)
    mask = ( x_list < 1e-6)
    result[mask] = 1 - x_list[mask]/3
    return np.array(result)
```

```python
plt.plot(x, F0(x))
plt.plot(x, 1 - (1./3.)*x)
```
