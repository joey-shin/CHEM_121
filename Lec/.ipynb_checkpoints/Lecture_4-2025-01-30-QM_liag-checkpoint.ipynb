## Quantum Mechanics and Linear Algebra

### **Wave Functions**
- **Definition**: A wave function $\psi(x,t)$ represents the quantum state of a particle.
- **Normalization**:
  $
  \int |\psi(x,t)|^2 \, dx = 1
  $
  
- Bra-ket notation: $ |\psi\rangle $
- **Superposition**:
  $
  | \psi(x,t)\rangle = c_1 |\psi_1(x,t)\rangle + c_2 |\psi_2(x,t)\rangle
  $



###  **Time-Independent Schr√∂dinger Equation**:
  $
  \hat{H} |\psi(x)\rangle = E |\psi(x)\rangle
  $
  
### **Momentum and Energy Operators**
- **Momentum Operator**:
  $
  \hat{p} = -i\hbar \frac{\partial}{\partial x}
  $
- **Hamiltonian Operator**:
  $
  \hat{H} = -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + U(x)
  $



###  **Matrix Representation**
- **State Vectors and Operator Matrices**:
  States are vectors $ |\psi\rangle $, and operators are matrices $ \hat{A} $.
- **Basis States**:
  $
  |\psi\rangle = \sum_i c_i |i\rangle
  $
- **Operator Matrix Elements**:
  $
  A_{ij} = \langle i|\hat{A}|j\rangle
  $
- **Full Matrix Representation**:
  Consider a simple two-state system with basis states $ |1\rangle $ and $ |2\rangle $. An operator $ \hat{A} $ might be represented as:
  $
  \hat{A} = \begin{bmatrix}
  \langle 1|\hat{A}|1\rangle & \langle 1|\hat{A}|2\rangle \\
  \langle 2|\hat{A}|1\rangle & \langle 2|\hat{A}|2\rangle
  \end{bmatrix}
  $
  If $ \hat{A} $ were the Hamiltonian $ \hat{H} $ in a simple model, its matrix elements could involve kinetic and potential energy terms specific to each state.
- **Diagonalization**:
  Diagonalizing the Hamiltonian matrix $ \hat{H} $:
  $
  \begin{bmatrix}
  H_{11} & H_{12} \\
  H_{21} & H_{22}
  \end{bmatrix}
  $
  leads to eigenvalues (energy levels) and eigenvectors (quantum states).


## Fourier Series Expansion

The complex Fourier series of a function \( f(t) \) with period \( T \) is given by:

$
f(t) = \sum_{n=-\infty}^\infty c_n e^{i 2\pi n t / T}
$

where \( c_n \) are the complex Fourier coefficients, calculated by:

$
c_n = \frac{1}{T} \int_0^T f(t) e^{-i 2\pi n t / T} dt
$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time period and create the time array
period = 1
time = np.linspace(0, period, 400, endpoint=False)

# Create the square wave function
y = np.sign(np.sin(2 * np.pi * time / period))

def cn(n):
    # Compute the Fourier coefficients
    c = y * np.exp(-1j * 2 * n * np.pi * time / period)
    return c.sum() / c.size

def f(x, Nh):
    # Fourier series approximation
    f = np.array([2 * cn(i) * np.exp(1j * 2 * i * np.pi * x / period) for i in range(1, Nh + 1)])
    return f.sum()



# Plotting the original and the approximated functions
plt.figure(figsize=(10, 5))
plt.plot(time, y, label='Original Square Wave')

for n_basis in [2, 5, 10, 15, 20, 50]:
    # Compute the Fourier series approximation
    y2 = np.array([f(t, n_basis).real for t in time])
    plt.plot(time, y2, label="No. basis: "+str(n_basis), linestyle='--')

plt.title('Square Wave and Its Fourier Series Approximation')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

### Basis set expansion
$$
\psi(x) = \sum_\mu c_\mu \phi_\mu (x)
$$

$\{ \phi_1, \phi_2, \ldots\}$ defines a basis set

$\{ c_1, c_2, \ldots\}$ are the corresponding expansion coefficients

###  **Time-Independent Schr√∂dinger Equation in basis set**:

$$
\hat{H} \sum_\mu c_\mu \phi_\mu (x) = E \sum_\mu c_\mu \phi_\mu (x)
$$

Multiply both sides by a bra:
$$
\int dr \phi_\nu^* (x) \hat{H} \sum_\mu c_\mu \phi_\mu (x) = E \int dr \phi_\nu^* (x) \sum_\mu c_\mu \phi_\mu (x)
$$

$$
\sum_\mu c_\mu \int dr \phi_\nu^* (x) \hat{H}  \phi_\mu (x) = E \sum_\mu c_\mu \int dr \phi_\nu^* (x)  \phi_\mu (x)
$$

In Bra-Ket notation:

$$
\sum_\mu \langle\psi_\nu| H |\psi_\mu\rangle c_\mu = E \sum_\mu \langle\psi_\nu|\psi_\mu \rangle c_\mu
$$

### The Schr√∂dinger Equation in matrix multiplication form
$H$ is a matrix, and its element

$H_{\nu\mu}$: $\langle\psi_\nu| H |\psi_\mu\rangle$

$S$ is a matrix, and its element

$S_{\nu\mu}$: $\langle\psi_\nu| \psi_\mu\rangle$

$$
\sum_\mu H_{\nu\mu} c_\mu = E \sum_\mu S_{\nu\mu} c_\mu
$$

$c$ is a vector with elements $c_\mu$

So we made the Schr√∂dinger Equation into a matrix multiplication form!

$$
H c = E S c
$$

#### If the basis set is orthonormal:
- Orthogonal means: $ S_{\nu\mu} = 0$ if $\nu \neq \mu$
- Normal means: $S_{\mu\mu} = 1$

In other words:
$S_{\nu\mu} = \delta_{\nu\mu}$ or $S = I$

Then:
$$Sc = Ic = c$$

and

$$Hc = Ec$$

This is an eigenvalue problem!

#### If the basis set is orthonormal:

(See Szabo Sec 3.4.5)

$$
(S^{-1}H) c = E c
$$

It is still an eigenvalue problem!

Note that $(S^{-1}H)$ is not symmetrix. We want its right eigenvectors

### Variational principle in quantum mechanics

The variational principle is a powerful method in quantum mechanics used to approximate the ground state energy of a system. It states that for any normalized trial wave function $\psi$ not orthogonal to the true ground state, the expectation value of the Hamiltonian $\hat{H}$, denoted as $E[\psi]$, provides an upper bound to the true ground state energy $E_0$. This principle is mathematically expressed as:

$$
E[\psi] = \frac{\langle \psi | \hat{H} | \psi \rangle}{\langle \psi | \psi \rangle} \geq E_0
$$


### Proof of the Variational Principle in Quantum Mechanics

The variational principle asserts that the expectation value of the Hamiltonian $\hat{H}$, computed using any trial wave function $\psi$ that is normalized and not orthogonal to the true ground state, will always be greater than or equal to the true ground state energy $E_0$. Here's how this principle is proved:

#### Setup:
Let $\alpha_0$ be the true ground state of the Hamiltonian $\hat{H}$ with energy $E_0$, and let $\psi$ be any normalized trial wave function. The trial wave function can be expressed as a linear combination of the orthonormal eigenstates $\alpha_n$ of $\hat{H}$, where $\hat{H} \alpha_n = E_n \alpha_n$ and $E_n$ are the corresponding eigenvalues.

$$
\psi = b_0 \alpha_0 + \sum_{n\neq 0} b_n \alpha_n
$$
where $b_n$ are coefficients such that $\sum_n |b_n|^2 = 1$ (normalization condition).

#### Expectation Value of $\hat{H}$:
The expectation value $E[\psi]$ of the Hamiltonian with respect to $\psi$ is given by:
$$
E[\psi] = \langle \psi | \hat{H} | \psi \rangle
$$
Substituting the expansion of $\psi$:
$$
E[\psi] = \left( b_0^* \alpha_0 + \sum_{m\neq 0} b_m^* \alpha_m \right) \hat{H} \left( b_0 \alpha_0 + \sum_{n\neq 0} b_n \alpha_n \right)
$$
$$
E[\psi] = |b_0|^2 E_0 + \sum_{n \neq 0} |b_n|^2 E_n
$$
since $\hat{H} \alpha_n = E_n \alpha_n$ and using the orthonormality $\langle \alpha_m | \alpha_n \rangle = \delta_{mn}$.

#### Inequality:
Since $E_n \geq E_0$ for all $n \neq 0$, and $\sum_n |b_n|^2 = 1$,
$$
E[\psi] = |b_0|^2 E_0 + \sum_{n \neq 0} |b_n|^2 E_n \geq |b_0|^2 E_0 + \sum_{n \neq 0} |b_n|^2 E_0 = E_0 (|b_0|^2 + \sum_{n \neq 0} |b_n|^2) = E_0
$$

#### Conclusion:
Thus, $E[\psi] \geq E_0$. This proves the variational principle, stating that the expectation value of the Hamiltonian for any normalized trial wave function is an upper bound to the ground state energy. This holds true as long as $\psi$ is not orthogonal to the true ground state, meaning $b_0 \neq 0$.


#### Key Steps in Applying the Variational Principle:

1. **Choose a Trial Wave Function**: The trial wave function $\psi$ should be a reasonable guess that incorporates as much knowledge about the system as possible. It often includes variational parameters that can be adjusted.

$$
\psi(x) = \sum_\mu c_\mu \phi_\mu (x)
$$

$\{ \phi_1, \phi_2, \ldots\}$ defines a basis set.

$\{ c_1, c_2, \ldots\}$ are the corresponding expansion coefficients satisfying the normialization condition:
$$
1 = \langle \psi(x) | \psi(x) \rangle = 
\sum_{\nu,\mu} c_\nu^* \langle  \phi_\nu(x) | \phi_\mu(x) \rangle c_\mu = \sum_{\nu,\mu} c_\nu^* S_{\nu\mu} c_\mu
= cSc
$$

2. **Compute the Expectation Value**: Calculate the expectation value of the Hamiltonian $\hat{H}$ using the trial wave function. 

$$
E[\psi] = \langle \psi(x) | \hat{H} | \psi(x) \rangle
= \sum_{\nu,\mu} c_\nu^* \langle  \phi_\nu(x) | \hat{H} | \phi_\mu(x) \rangle c_\mu
= \sum_{\nu,\mu} c_\nu^* H_{\nu\mu} c_\mu
= c H c
$$


3. **Minimize the Energy**: Adjust the variational parameters in the trial wave function to minimize the expectation value $E[\psi]$. The minimum value obtained through this process is an approximation to the ground state energy $E_0$.

### Minimizing Functions with Constraints

Minimizing a function subject to constraints is a common problem in various fields including physics, economics, and engineering. The method of Lagrange multipliers is a powerful technique for finding the local minima or maxima of a function subject to equality constraints.

#### Concept of Lagrange Multipliers

The basic idea is to transform a constrained optimization problem into a problem without constraints by introducing additional variables called Lagrange multipliers.

#### Problem Setup

Consider a function $f(\mathbf{x})$ that you want to minimize or maximize subject to a constraint $g(\mathbf{x}) = 0$. Here, $\mathbf{x}$ represents a vector of variables $(x_1, x_2, ..., x_n)$.

#### Lagrangian Function

The Lagrangian $\mathcal{L}$ is defined as:
$$
\mathcal{L}(\mathbf{x}, \lambda) = f(\mathbf{x}) - \lambda (g(\mathbf{x}) - c)
$$
where $\lambda$ is the Lagrange multiplier, and $c$ is the constant to which the constraint function $g$ is set, often zero for simplicity.

#### Solving the Problem

1. **Form the Lagrangian**: Write down the Lagrangian incorporating all constraints with their respective multipliers.
2. **Find the Derivatives**: Compute the partial derivatives of $\mathcal{L}$ with respect to each variable $x_i$ and the Lagrange multiplier $\lambda$.
3. **Set Equations to Zero**:
   - The conditions for a minimum or maximum are given by:
     $$
     \frac{\partial \mathcal{L}}{\partial x_i} = 0, \quad \frac{\partial \mathcal{L}}{\partial \lambda} = 0
     $$
4. **Solve the System of Equations**: The above steps give a system of equations which, when solved, yield the values of $\mathbf{x}$ and $\lambda$ that minimize or maximize $f$ under the constraint $g(\mathbf{x}) = c$.

#### Example

If you want to minimize the function $f(x, y) = x^2 + y^2$ subject to the constraint $x + y = 1$, the Lagrangian is:
$$
\mathcal{L}(x, y, \lambda) = x^2 + y^2 - \lambda (x + y - 1)
$$
Solving $\frac{\partial \mathcal{L}}{\partial x} = 0$, $\frac{\partial \mathcal{L}}{\partial y} = 0$, and $\frac{\partial \mathcal{L}}{\partial \lambda} = 0$ provides the values of $x$, $y$, and $\lambda$ that minimize $f(x, y)$ with the given constraint.


### Constrained optimization of the trial wave function

The Lagrangian is:
$$
\mathcal{L}(\psi) = cHc - \lambda c S c = \sum_{\nu,\mu} c_\nu^* (H_{\nu\mu}-\lambda S_{\nu\mu}) c_{\mu}
$$

The derivative is
$$
\dfrac{\partial \mathcal{L}}{c_\nu} = 2 \sum_{\mu} (H_{\nu\mu}-\lambda S_{\nu\mu}) c_\mu
$$

Setting the derivative to zero:
$$
 H c = \lambda S c
$$

What value of $\lambda$ ensures normalization? ($cSc=1$)
$$
c H c =  \lambda (c S c) = \lambda
$$
That is $E$!
