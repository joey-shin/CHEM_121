# Question 1

## (i)

Setting $\epsilon = \hbar^2/2m$ will cancel out the constants in the kinetic energy term of the Hamiltonian, reducing it to just the second order derivative operator. 
$$
\hat{H}^* = \frac{\hat{H}}{\epsilon} = \frac{\left(-\dfrac{\hbar^2}{2m} \dfrac{d^2}{dx^2}\right)}{\dfrac{\hbar^2}{2m}} = - \frac{d^2}{dx^{*2}}
$$

## (ii)

Starting with the Time-Independent Schrödinger Equation
$$
\hat{H^*} \psi_n(x^*) = E_n \psi_n(x^*)
$$
Substituting 
$
\hat{H} = -\frac{d^2}{dx^{2}}
$
and
$
\psi_n(x) = A\sin(n\pi x)
$
into the Schrödinger Equation
$$
 \left[-\frac{d^2}{dx^{2}}\right] \psi_n(x) = E_n \psi_n(x)
$$
Evaluating
$
\frac{d^2}{dx^2} A\sin(n\pi x)
$
:

$$
 \frac{d^2}{dx^2} A\sin(n\pi x) = n\pi A\frac{d}{dx} \cos(n\pi x) = -n^2\pi^2 A\sin(n\pi x)
$$
$$
 \frac{d^2}{dx^2} \psi_n(x) = -n^2\pi^2 \psi_n(x)
$$
Substituting this result back into the Schrödinger Equation
$$
-\left(-n^2\pi^2 \psi_n(x) \right) = E_n \psi_n(x)
$$
$$
 \pi^2 n^2 \psi_n(x) = E_n \psi_n(x)
$$
$$
 \pi^2 n^2 = E_n
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def energy(n):
    return n**2 * np.pi**2

def planewave(x, n, A):
    return A * np.sin(n*np.pi*x)
```

```python
ns = range(1, 5)
A = 10
x = np.linspace(0, 1, 500)

fig, ax = plt.subplots()
ax.axvline(x=0, color='black') # verticle line at x=0
ax.axvline(x=1, color='black') # verticle line at x=1
for n in ns:
    ax.plot(x, planewave(x, n, A) + energy(n), label=f'n={n}')
    ax.axhline(y=energy(n), color='black', ls='--') # horizontle lines including E = 0
    ax.text(1.07, energy(n), f'$E_{n}$', verticalalignment='center')
ax.set_title('1D Particle in a Box')
ax.set_xlabel('X')
ax.set_ylabel('Energy')
ax.legend(loc='upper right')
ax.grid()
plt.savefig('Visualizations/1D_Particle_in_a_Box.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_space = np.linspace(-5, 5, 500)
Ux = 0.5 * x_space**2. #defining dimentionless potential well
psi0_exact = np.pi**(-1./4.) * np.exp(-x_space**2./2) #exact solution for ground state

# Create the figure and first axes
fig, ax1 = plt.subplots()
ax1.plot(x_space, Ux, 'k--', label='U(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('U(x)')
ax1.legend()

# Create a second y-axis on the right
ax2 = ax1.twinx()
ax2.plot(x_space, psi0_exact, 'g-', label='$\psi$')
ax2.set_ylabel('$\psi_0$ Exact')
ax2.legend()
```

```python
def phi_A_func(x_space, A, alpha=2.): #defining gaussian basis set
    return np.exp(-alpha*(x_space - A) ** 2.)

def S_AB_func(xA, xB, alpha): # analytical integral of Overlap matrix elements
    return (0.5*np.pi/alpha)**0.5 * np.exp(-0.5*alpha* (xA - xB)**2 )

def H_AB_func(xA, xB, alpha): # analytical integral of Hamiltonian matrix elements
    return 0.5* S_AB_func(xA, xB, alpha) * (alpha - alpha**2 * (xA - xB)**2 + \
                             0.25*(1/alpha + (xA + xB)**2 ))

def S_H_AB_func(xA, xB, alpha): # combining results of both overlap integral expectation values
    S_AB =  S_AB_func(xA, xB, alpha)
    H_AB = 0.5* S_AB * (alpha - alpha**2 * (xA - xB)**2 + \
                             0.25*(1/alpha + (xA + xB)**2 ))
    return S_AB, H_AB

def find_eigval_eigvec(S_AB, H_AB): # computing the eigenvector and eigenvalues of S^{-1} @ H, where the eigen values are the energies and the coefficients are the eigenstates
    # compute S^{-1}H
    S_inv_H = np.linalg.inv(S_AB) @ H_AB
    # eigval[i]  -> eigenvectors[:,i]
    eigval, eigvec = np.linalg.eig(S_inv_H)
    
    index_sort = np.argsort(eigval)
    eigval_sorted = eigval[index_sort]
    eigvec_sorted = eigvec[:, index_sort]
    
    for i in range(len(eigvec_sorted)):
        eigvec_sorted[:,i] /= np.sqrt(eigvec_sorted[:,i] @ S_AB @ eigvec_sorted[:,i])
        
    return eigval_sorted, eigvec_sorted
```

```python
alpha = 2 # setting alpha exponent to 2 

gaussian_centers = np.linspace(-4, 4, 9) # expanding number of basis sets to 9 with centers -4 to 4
n_basis = len(gaussian_centers)

n_basis
```

```python
eigval, eigvec = {}, {}

for n_basis in [5, 9, 15, 21]: # solution to the schrodinger equation with varied basis lengths
    gaussian_centers = np.linspace(-4, 4, n_basis)
    S_AB, H_AB = S_H_AB_func(gaussian_centers[:,None], gaussian_centers[None,:], alpha)
    eigval[n_basis], eigvec[n_basis] = find_eigval_eigvec(S_AB, H_AB)
```

```python
plt.plot(np.arange(20), np.arange(20)+0.5, '--', label='True') # true energies of harmonic oscillator potential
for n_basis in [5, 9, 15, 21]: # plot of the accuracy of each eigenenergies corresponding to each eigenstates for the different number of bases
    plt.plot(np.arange(len(eigval[n_basis])), eigval[n_basis], 'o-', label='No basis '+str(n_basis))
plt.xlabel("No of states")
plt.ylabel('energy (reduced units)')
plt.legend()
```

```python
Psi = {}

wave_function_index = 4 # choosing the 4th excited state of the harmonic oscillator

for n_basis in [5, 9, 15, 21]: 
    Psi[n_basis] = np.zeros_like(x_space)
    gaussian_centers = np.linspace(-4, 4, n_basis)
    for i, g_A in enumerate(gaussian_centers):
        # print(n_basis, eigvec[n_basis].shape)
        Psi[n_basis] += phi_A_func(x_space, g_A, alpha) * eigvec[n_basis][i,wave_function_index] #generating the y-values of the 4th excited state wavefunctionwith linear combinatoins of basis vectors with its corresponding coefficients for all basis lengths 
```
